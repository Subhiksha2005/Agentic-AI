```python
import os
import random
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.tools import Tool
from langchain.agents import initialize_agent, AgentType
```

```python
os.environ["GOOGLE_API_KEY"] = "AIzaSyAkQLZvX5bqDoRWf5R6htdFXP9CleL8sH4"

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0.6
)
```

```python
workout_tool = Tool(
    name="Workout Planner Tool",
    func=lambda x: """
🏋️ 7-Day Workout Plan:

Day 1: Chest + Triceps
Day 2: Back + Biceps
Day 3: Legs
Day 4: Rest
Day 5: Shoulders
Day 6: Cardio + Abs
Day 7: Full Body Stretch
""",
    description="Use this tool to generate workout plan"
)
```

```python
diet_tool = Tool(
    name="Diet Planner Tool",
    func=lambda x: """
🥗 Simple Healthy Diet Plan:

Breakfast: Oats + Fruits
Lunch: Rice + Vegetables + Protein
Snack: Nuts / Smoothie
Dinner: Light meal + Salad
Drink 3L water daily
""",
    description="Use this tool to generate diet plan"
)
```

```python
health_tool = Tool(
    name="Health Tips Tool",
    func=lambda x: """
💧 Daily Health Tips:

• Sleep 7-8 hours
• Avoid junk food
• Walk 8,000+ steps
• Stretch daily
• Reduce screen time
""",
    description="Use this tool to provide health tips"
)
```

```python
motivation_tool = Tool(
    name="Motivation Booster Tool",
    func=lambda x: random.choice([
        "🔥 Your body can do it. It's your mind you must convince.",
        "💪 Discipline equals freedom.",
        "🚀 Progress, not perfection.",
        "🌟 Healthy habits create a healthy life."
    ]),
    description="Use this tool to motivate users"
)
```

```python
agent = initialize_agent(
    tools=[workout_tool, diet_tool, health_tool, motivation_tool],
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)
```

```python
response = agent.invoke({
    "input": "Create a weekly fitness and diet plan for beginners."
})

print(response["output"])
```
